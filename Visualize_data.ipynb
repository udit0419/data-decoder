```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
```

```python
df = pd.read_csv('fifa21_male2.csv')
print(df.head())
print(df.info())
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['OVA'], bins=30, kde=True, color='lightcoral')
plt.title('Distribution of Overall Rating')
plt.xlabel('market value (EUR)')
plt.ylabel('Frequency')
plt.xscale('log')
plt.grid(True)
plt.show()
```

```python
top_clubs = df['Club'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_clubs.values, y=top_clubs.index, palette='viridis')
plt.title('Top 10 Clubs with the Most Players')
plt.xlabel('Number of players')
plt.ylabel('Club')
plt.grid('true')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Position', y='Wage')
plt.title('Wage Distribution by Player Positions')
plt.xlabel('Player Positions')
plt.ylabel('Wage (EUR)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
